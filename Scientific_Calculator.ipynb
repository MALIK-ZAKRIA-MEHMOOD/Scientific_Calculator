```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Functions for scientific operations
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    else:
        return x / y

def power(x, y):
    return x ** y

def sqrt(x):
    return math.sqrt(x)

def log(x):
    if x > 0:
        return math.log(x)
    else:
        return "Error! Logarithm is only defined for positive numbers."

def sin(x):
    return math.sin(math.radians(x))

def cos(x):
    return math.cos(math.radians(x))

def tan(x):
    return math.tan(math.radians(x))

# Function to plot graphs
def plot_function(func_name):
    x = np.linspace(-360, 360, 400)  # Generate 400 points between -360 to 360 degrees
    plt.figure(figsize=(8, 5))

    if func_name == 'sin':
        y = np.sin(np.radians(x))
        plt.plot(x, y, label="sin(x)")
    elif func_name == 'cos':
        y = np.cos(np.radians(x))
        plt.plot(x, y, label="cos(x)")
    elif func_name == 'tan':
        y = np.tan(np.radians(x))
        plt.plot(x, y, label="tan(x)")
    elif func_name == 'log':
        x = np.linspace(0.1, 10, 400)  # Logarithms are only defined for positive values
        y = np.log(x)
        plt.plot(x, y, label="log(x)")
    elif func_name == 'x^2':
        y = x**2
        plt.plot(x, y, label="x^2")
    else:
        print("Invalid function")

    plt.title(f"Graph of {func_name}(x)")
    plt.xlabel("x-axis")
    plt.ylabel(f"{func_name}(x)")
    plt.grid(True)
    plt.legend()
    plt.show()

# User interaction loop
while True:
    print("\n--- Scientific Calculator ---")
    print("Choose an operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Power (x^y)")
    print("6. Square Root (√x)")
    print("7. Logarithm (ln x)")
    print("8. Sine (sin x)")
    print("9. Cosine (cos x)")
    print("10. Tangent (tan x)")
    print("11. Plot Graph")
    print("0. Exit")

    choice = input("Enter your choice: ")

    if choice == '0':
        break

    elif choice in ['1', '2', '3', '4', '5']:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == '1':
            print(f"Result: {add(num1, num2)}")
        elif choice == '2':
            print(f"Result: {subtract(num1, num2)}")
        elif choice == '3':
            print(f"Result: {multiply(num1, num2)}")
        elif choice == '4':
            result = divide(num1, num2)
            print(f"Result: {result}")
        elif choice == '5':
            print(f"Result: {power(num1, num2)}")

    elif choice == '6':
        num = float(input("Enter a number: "))
        print(f"Square Root of {num}: {sqrt(num)}")

    elif choice == '7':
        num = float(input("Enter a positive number: "))
        result = log(num)
        print(f"Logarithm of {num}: {result}")

    elif choice == '8':
        angle = float(input("Enter angle in degrees: "))
        print(f"Sine of {angle} degrees: {sin(angle)}")

    elif choice == '9':
        angle = float(input("Enter angle in degrees: "))
        print(f"Cosine of {angle} degrees: {cos(angle)}")

    elif choice == '10':
        angle = float(input("Enter angle in degrees: "))
        print(f"Tangent of {angle} degrees: {tan(angle)}")

    elif choice == '11':
        graph_choice = input("Enter the function to plot (sin, cos, tan, log, x^2): ")
        plot_function(graph_choice)

    else:
        print("Invalid choice. Please try again.")
```
